```python
# filter waves
# imp1 = df[(df['wave']>=10) | (df['wave']<=5)]
imp1 = df.copy()
```

```python
imp1.describe()
```

```python
# filter columns
imp2 = imp1[id_cols + hob_cols + rat_cols + oth_cols]
imp2.head()
```

```python
na_cols = []
for col in imp2.columns:
    if imp2[col].isna().sum() > 0:
        na_cols.append(col)
    print(col, " has ", str(imp2[col].isna().sum()), " NA values.")
```

```python
imp3 = imp2.copy()
print('remove ', len(imp3[imp3['fun5_1'].isna()]), ' records')
# print(len(imp2))
imp3 = imp3[~imp2['fun5_1'].isna()]
print(f'df now with length {len(imp3)}')
```

```python
imp3.isna().sum()
```

```python
print('remove date.isna(), one participant')
imp3 = imp3[~imp3['date'].isna()]
```

```python
print('fill shar1_1 with 0 as 100 points are given to other attributes')
imp3['shar1_1'] = np.where(imp3['shar1_1'].isna(), 0, imp3['shar1_1'])
```

### group by individual iid

```python
imp3['iid'].nunique()
```

```python
imp4 = imp3.copy()
imp4 = imp4.groupby('iid').agg('max')
imp4.drop(['pid', 'match'], axis=1, inplace=True)
```
